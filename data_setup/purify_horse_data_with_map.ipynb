```python
import numpy as np
import pandas as pd
import dask.dataframe as dd
from multiprocessing.dummy import Pool
import os
import glob

pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
```

```python
horse_df = pd.read_csv("../data/csv/horse_data/horse_63/1978105785.csv", sep = ",")

def readcsv_map_multi(filelist):
    p = Pool(os.cpu_count())
    df = pd.concat(p.map(pdreadcsv, filelist))
    p.close()
    return df

def pdreadcsv(csv_path):
    return pd.read_csv(csv_path)

for i in range(64):
    files = glob.glob(os.path.join("../data/csv/horse_data/horse_" + str(i), '*.csv' ))
    horse_tmp_df = readcsv_map_multi(files)
    horse_df = pd.concat([horse_df, horse_tmp_df], axis = 0)
   
```

win_count

```python
horse_character_df = pd.DataFrame(columns = ["win_count"])
horse_character_df["win_count"] = horse_df.groupby(["horse_id"], as_index = True).apply(lambda d: (d.order == '1').sum())

horse_featured_df = pd.merge(horse_featured_df, horse_character_df, on = 'horse_id')
```

update_total_race_number

```python
update_total_race_number_df = pd.DataFrame(columns = ["update_total_race_number"])
update_total_race_number_df["update_total_race_number"] = horse_df.groupby(["horse_id"], as_index = True).apply(lambda d: (d.order == '1').count())

horse_featured_df = pd.merge(horse_featured_df, update_total_race_number_df, on = 'horse_id')
```

win_count_grass

```python
win_count_grass_df = pd.DataFrame(columns = ["win_count_grass"])
win_count_grass_df["win_count_grass"] = horse_df.groupby(["horse_id"], as_index = True).apply(lambda d: ((d.order == '1') & (d.ground_type_芝 == 1)).sum())

horse_featured_df = pd.merge(horse_featured_df, win_count_grass_df, on = 'horse_id')
```

win_count_dart

```python
win_count_dart_df = pd.DataFrame(columns = ["win_count_dart"])
win_count_dart_df["win_count_dart"] = horse_df.groupby(["horse_id"], as_index = True).apply(lambda d: ((d.order == '1') & (d.ground_type_ダ == 1)).sum())

horse_featured_df = pd.merge(horse_featured_df, win_count_dart_df, on = 'horse_id')
```

race_count_grass
