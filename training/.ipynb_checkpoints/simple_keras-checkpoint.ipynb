```python
# label to integral 
# nomalization 

#labels_class = LabelEncoder()
#y = labels_class.fit_transform(labels["order"])

row_y = labels["order"].values
y = np_utils.to_categorical(row_y)

standard_scale = StandardScaler()
x = standard_scale.fit_transform(row_x)

```

```python
# win5 data
win5_row_y = win5_labels["order"].values
win5_race_id_row = win5_labels["race_id"].values #全レースあるわけではない
win5_y = np_utils.to_categorical(win5_row_y)

standard_scale = StandardScaler()
win5_x = standard_scale.fit_transform(win5_row_x)
print(win5_x.shape)
```

```python
# split train test data

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
#pd.DataFrame(X_train).to_csv("csv/data/X_train.csv", index = False)
#pd.DataFrame(X_test).to_csv("csv/data/X_test.csv", index = False)
#pd.DataFrame(Y_train).to_csv("csv/data/Y_train.csv", index = False)
#pd.DataFrame(Y_test).to_csv("csv/data/Y_test.csv", index = False)
```

```python
# make model
OWN_FILE_NAME = path.splitext(path.basename('\\Users\\vmlab\\win5.ext'))[0]

# weights = [13.380822, 13.315343, 13.474366, 13.581180, 13.835053, 13.783522, 13.862002, 14.299073, 14.626877, 15.455229, 16.458815, 18.265548, 20.454344, 23.344688, 27.501486, 35.942113, 159.234079, 215.151163]
model = Sequential()
model.add(Dense(256, kernel_regularizer = regularizers.l2(0.001), activation = 'relu', input_dim = X_train.shape[1]))
model.add(Dropout(0.5))
model.add(Dense(1028, kernel_regularizer = regularizers.l2(0.001), activation = 'relu'))
model.add(Dropout(0.3))
model.add(Dense(512, kernel_regularizer = regularizers.l2(0.001), activation = 'relu'))
model.add(Dense(19, activation = 'softmax'))      # output is a number of order

model.compile(loss = keras.losses.categorical_crossentropy, optimazer = 'adam', metrics = ['accuracy'])

history = model.fit(X_train, Y_train, batch_size = 128, epochs = 50, validation_split = 0.2)
model.save("model/{}_simple_model_others.h5".format(OWN_FILE_NAME))
```

```python
# predict

pred = model.predict(X_test)
pred_order = np.argmax(pred, axis = 1)
Y_ans = np.argmax(Y_test, axis = 1)

#print(pred[248][1]*100)
#print(pred_order[248])
#print(Y_ans[248])
```
